## 01.Wstęp

Wybrałem hashtag #Brexit , czyli wyjście Wielkiej brytani z uni europejskiej. Dane zostały pobrane za pomocą snscrape,
w 3 przedziałach czasowych: 

-Przed brexit 
   
-Tuż po wyjsciu z ue 
   
-aktualnie

## 02.Preprocessing

Dane zapisane w plikach, wyciągnałem treści tweetów  oraz daty, dla lepszych wyników usunałem znak # . Przekazałem treści do dwóch funckji: text2emotion oraz vader od nltk

Wiekszy preprocessing zastosowałem do utworzenia chmur tagów.

Połączyłem wszystkie tweety, usunałem niepotrzebne cudzysłowia, interpunkcje. 

Oprócz stopwords z bibioteki dodtałem :     custom_stop_words = ['https','brexit', "http", 'uk', 'amp', 'nt', 'u', 'nn','n','de','eu']
na końcu lematyzacja

## 03.Analiza

Podzieliłem analize na 2 tematy, Analiza całościowa oraz Analiza czasowa.

### 01.Analiza Czasowa


Tak jak wcześniejsz wspomniałem pobrałem tweety z 3 wymirów czasowych. Wykresy pokazują średnie nastawienie dla danych dni.

Pierwszy wykres to ilustracja prze Brexitem. Widzimy, żę nastawianie ludzi się wacha lech pod sam koniec , czyli styczeń 2020 opinia jest zasadniczo lepsza. Jest tak gdyż 01.luty.2020 Wielka Brytania wyszła z Uni Europejskiej 

Na nastepnym wykresie widzimy przedstawienie opini odrazu po Brexit. Tak jak wyżej w czasie opuszczenia blisko 1 lutego opinia ludzi jest pozytywn, natomiast z biegiem czasu widzimy wiecej negatywny wypowiedzi 

Ostatni wykres w tym podrozdziale przedstawia aktualne opinie o Brexit. Odrazu widać, żę raczej ludzie nie są zadowoleni z tego ruchu Wielkiej Brytani, są jakieś pozytywne opinie natomiast ale nie za wiele

#### Podsumowanie
Można zauważyć, że opinie z biegiem czasu stały sie mniej neutralne oraz bardziej negatywne. Czyli ludzie na tweeter nie uwielbia, żę Wielka Brytania wyszła z ue. Ciekawe, żę styczeń-luty 2020 , wtedy jak GB formalnie opusciło EU, na tweeter były zdecydowanie wiecej pozytywnych wpisów. Aktualnie wiecej negetywnych tweetów , zwiazane to z wysokimi cenami

## 04.Podsumowanie
Obserwujemy, żę przed Brexit zdania były podzielone. Styczeń/luty 2020 wtedy gdy brexit dobiegał do skótku, opinie były pozytywne (można zobaczyć na wykresie ). Natomiast póżniej jest wiecej negatywnych wpisów. Jeśli chodzi o konkretne emocje można wywioskować o co chodziło. Często pojawiał się borisjohnson czyli były premier Wielkiej Brytani, oraz po Brexit pojawiało się "Tory" nawiazuję do Partii Konserwatywnej, znana również jako Tories. Ciekawe, że pojawiło się raz "trump" , możliwe , że trump miał jakaś kontrowersyjna wypowiedź
